```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

df = pd.read_csv(r'freiotto02.txt', sep=" ", skiprows=3, header=None)
dfnew = df.rename(columns={0: "type", 1: "x", 2: "y", 3: "z" })
```

```python
vPos = dfnew.loc[dfnew['type'] == 'v']
print(vPos)
vPosSort = vPos.sort_values(by=['z']) # "x" works
vPosClip = vPosSort.iloc[:46866] #max is 46866
print(vPosClip)
```

```python
fig = plt.figure()
fig.set_size_inches(30, 15)
ax = fig.add_subplot(111, projection='3d', proj_type = 'ortho')
#ax.view_init(30, 100)

x = vPosClip['x'].values
y = vPosClip['z'].values
z = vPosClip['y'].values

X, Y, Z = np.array([x]), np.array([y]), np.array([z])

ax.scatter(x, y, z, c='r', marker='o', s=0.5)
ax.set_xlabel('x')
ax.set_ylabel('z')
ax.set_zlabel('y')
ax.view_init(15, 90) #30, 100 seem to be good vals

plt.show()
```
